```
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sn
```

```
df=pd.read_csv("/content/PhiUSIIL_Phishing_URL_Dataset.csv")
```

```
df.head()
```

```
df.dropna(inplace=True)
print("Null values after dropping rows:")
print(df.isnull().sum())
```

```
df.head()
```

```
print("Original data type of 'label' column:", df['label'].dtype)
```

```
df.head()
```

```
df['label'] = df['label'].astype(int)
print("Data type of 'label' column after conversion:", df['label'].dtype)
```

```
df.head()
```

```
import re


def count_special_chars(url):

    return len(re.findall(r'[^a-zA-Z0-9/:.?=_\-]', url))


df['NoOfSpecialChars'] = df['URL'].apply(count_special_chars)
df['HasWWW'] = df['Domain'].apply(lambda x: 'www' in str(x).lower())

print("New features 'NoOfSpecialChars' and 'HasWWW' created.")
print(df[['URL', 'Domain', 'NoOfSpecialChars', 'HasWWW']].head())
```

```
plt.figure(figsize=(12, 5))


plt.subplot(1, 2, 1)
sns.countplot(x='label', data=df)
plt.title('Distribution of Label (0: Legitimate, 1: Phishing)')
plt.xlabel('Label')
plt.ylabel('Count')



plt.subplot(1, 2, 2)
sns.histplot(df['URLLength'], kde=True)
plt.title('Distribution of URL Length')
plt.xlabel('URL Length')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
print("Visualized distribution of 'label' and 'URLLength' columns.")
```
